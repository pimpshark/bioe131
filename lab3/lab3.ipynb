### Read in a text file, create a dictionary for with one entry for every unique word, followed by the number of words

```python
import string
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
all = []
with open('romeoandjuliet.txt', 'rt') as input:
    for i in input:
        all.append(i)
```

```python
text = ' '.join(''.join(all).replace('\n', ' ').split()).upper()
#text = text.replace('.','').replace(',','').replace("\'",'').replace(';','').replace('[','').replace(']',''')''')
#text.replace('(','').replace(')','').replace('!')
```

```python
exclude = set(string.punctuation)
text2 = ''.join(ch for ch in text if ch not in exclude)
```

```python
words = text2.split(' ')
```

```python
worddict = {}
for i in words:
    try:
        worddict[i] += 1
    except:
        worddict[i] = 1
        
```

### Make barplot from dictionary

```python
y = sorted(worddict.values(),reverse=True)
N = len(y)
x = range(N)
```

```python
width = .7
high = max(y)
#plt.ylim(0,700)
plt.bar(x, y, width, color="blue")
plt.rcParams['figure.figsize'] = (10,10)
plt.xlabel('words')
plt.ylabel('count')
plt.title('count of unique words')
plt.show()
```

Looks like, because of how thin the bars of the bar graph are, some bars are not visible, such as the bar for the word "and", which should have a word count over 700. 

```python
sorted(y,reverse=True)
sorted(worddict,key=worddict.__getitem__,reverse=True)
```

```python
sorted(y)
sorted(worddict,key=worddict.__getitem__,)
```

Looks like the most common words are "and, the, I, to, a, of, my, that, is, in" and the least common words are "wheels, hasten, adversitys, minded, grieve, meeting, unknown, smatter, mistaking, titans". This makes sense because the most common words are words that are extremely common in the english language in general, so they are the most common words for most english writers, including shakespeare. The least common words are words which are more specific, and so can represent either words which shakespeare personally doesn't like using, or words which were not incredibly useful for advancing the story. 

```python
dsort = sorted(worddict.values(),reverse=True)
rank=[]
for i in range(len(dsort)):
    rank.append(i+1)
```

```python
len(rank)
```

```python
plt.scatter(dsort,rank)
plt.title('Scatter plot count vs rank')
plt.xlabel('count')
plt.ylabel('rank')
plt.show()
```

```python
plt.scatter(np.log(dsort),np.log(rank))
plt.title('Scatter Plot log(count) vs log(rank)')
plt.xlabel('log(count)')
plt.ylabel('log(rank)')
plt.show()
```

These plots demonstrate that there is almost an exponential relationship between the number of times a word occurs and how common it is compared to the other words. One can also see that the log graph is not incredibly linear. This is because the most common words are those which are used incredibly often, significantly more than other words. On the other hand, there are many words that all have a count of 1, but all have different ranks. This can be seen in the log graph at the ends of the graph. However, ignoring these datapoints, one sees that the middle of the graph is still linear, which implies that there is an exponential relationship between word count and rank not dependent on most common words or the least common words. I believe that this relationship exists because as words become more common, the amount of times they are used increases at an exponential rate rather than a linear one.
